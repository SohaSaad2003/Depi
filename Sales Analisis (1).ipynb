```python
# Extract year, month, and day from ship_date
df['ship_year'] = df['ship_date'].dt.year
df['ship_month'] = df['ship_date'].dt.month
df['ship_day'] = df['ship_date'].dt.day
```

## Data Analysis Questions

#### What is the total sales per year?

```python
total_sales_per_year = df.groupby('order year')['sales'].sum()

total_sales_per_year
```

```python

total_sales_per_year = df.groupby('order year')['sales'].sum()


plt.figure(figsize=(6, 4))
total_sales_per_year.plot(kind='bar', color='skyblue')
plt.title('Total Sales Per Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

####  Which product category has the highest sales?

```python
sales_by_category = df.groupby('category')['sales'].sum().sort_values(ascending=False)

sales_by_category
```

```python
sales_by_category = df.groupby('category')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(6, 4))
sales_by_category.plot(kind='bar', color='lightcoral')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

highest_sales_category = sales_by_category.idxmax()
highest_sales_value = sales_by_category.max()

print(f"The product category with the highest sales is '{highest_sales_category}' with total sales of {highest_sales_value}.")
```

#### What is the distribution of sales across different regions?

```python
sales_by_region = df.groupby('region')['sales'].sum()

sales_by_region
```

```python
plt.figure(figsize=(8, 6))

sales_by_region = df.groupby('region')['sales'].sum()
plt.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))

plt.title('Sales Distribution Across Regions', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

#### What is the average shipping cost per shipping mode?

```python
avg_shipping_cost = df.groupby('ship mode')['shipping cost'].mean()

avg_shipping_cost
```

```python
plt.figure(figsize=(6, 4))

avg_shipping_cost = df.groupby('ship mode')['shipping cost'].mean().sort_values()

sns.barplot(x=avg_shipping_cost.index, y=avg_shipping_cost.values)

plt.title('Average Shipping Cost by Shipping Mode', fontsize=16)
plt.xlabel('Shipping Mode')
plt.ylabel('Average Shipping Cost')
plt.xticks(rotation=45)  
plt.tight_layout()        
plt.show()

```

#### How does discount affect profit?

```python
discount_profit_relation = df.groupby('discount')['profit'].mean()

discount_profit_relation
```

```python
plt.figure(figsize=(6, 4))

sns.scatterplot(x='discount', y='profit', data=df, hue='category', palette='Set1', s=100)

plt.title('Impact of Discount on Profit', fontsize=16)
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

#### Which regions are the most profitable?

```python
profit_by_region = df.groupby('region')['profit'].sum().sort_values(ascending=False)
profit_by_region
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=profit_by_region.index, y=profit_by_region.values, color='skyblue')  # Using a single color
plt.title('Total Profit by Region', fontsize=16)
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() 
plt.show()


most_profitable_regions = profit_by_region.head()  
print("The most profitable regions are:\n", most_profitable_regions)

```

#### What is the impact of order priority on sales performance?

```python
sales_by_priority = df.groupby('order priority')['sales'].mean()

sales_by_priority
```

```python
sales_by_priority = df.groupby('order priority')['sales'].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=sales_by_priority.index, y=sales_by_priority.values)
plt.title('Average Sales by Order Priority', fontsize=16)
plt.xlabel('Order Priority')
plt.ylabel('Average Sales')
plt.xticks(rotation=45) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() 
plt.show()

print("Average sales by order priority:\n", sales_by_priority)
```

####  How is the sales trend over time (monthly)?

```python
sales_trend = df.groupby(['order year', 'order month'])['sales'].sum().unstack()
sales_trend
```

```python
plt.figure(figsize=(10, 4))
sales_trend = df.groupby(['order year', 'order month'])['sales'].sum().reset_index()
sns.lineplot(x='order month', y='sales', hue='order year', data=sales_trend, marker="o")
plt.title('Monthly Sales Trend Over the Years', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 13))  # Ensure x-axis shows all months
plt.legend(title='Year', loc='upper right')
plt.grid(True)
plt.show()
```

#### What are the most ordered products?

```python
most_ordered_products = df.groupby('product name')['quantity'].sum().sort_values(ascending=False).head(10)

most_ordered_products
```

```python
most_ordered_products = df.groupby('product name')['quantity'].sum().sort_values(ascending=False).head(10)

most_ordered_products = most_ordered_products.reset_index()

plt.figure(figsize=(10,4))
sns.barplot(data=most_ordered_products, x='quantity', y='product name', palette='viridis')

plt.title('Top 10 Most Ordered Products')
plt.xlabel('Total Quantity Ordered')
plt.ylabel('Product Name')

plt.show()
```

#### What is the average profit per product category?

```python
avg_profit_by_category = df.groupby('category')['profit'].mean()
avg_profit_by_category
```

```python
avg_profit_by_category = df.groupby('category')['profit'].mean()
avg_profit_by_category = avg_profit_by_category.reset_index()

sns.set(style="whitegrid")

plt.figure(figsize=(10, 4))
sns.barplot(x='category', y='profit', data=avg_profit_by_category, palette='viridis')

plt.title('Average Profit by Category', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Average Profit', fontsize=14)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
